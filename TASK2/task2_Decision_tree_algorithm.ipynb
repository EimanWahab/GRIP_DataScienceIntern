### DECISION TREE CLASSIFIER
A Decision Tree is a machine learning model that recursively splits data based on features to make predictions.
It partitions data into branches of decisions, ultimately providing insights into classification or regression tasks.
Its tree structure visualizes decision-making in a hierarchical manner.

```python
# Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Create a Decision Tree Classifier
Decision_tree = DecisionTreeClassifier(random_state=42)
Decision_tree.fit(X, y)
```

```python
# Visualize the Decision Tree graphically
plt.figure(figsize=(12, 9))
tree.plot_tree(Decision_tree, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.title("Decision Tree Classifier - Iris Dataset")
plt.show()
```
